```
import numpy as np # biblioteca numérica do python
```

# Sum

realiza a **soma** de todos os elementos de uma lista um de um array

```
# sum 

a = [3,3,3]

print(3+3+3)
print(sum(a))
```

# np.mean

realiza a **média** de todos os elementos de uma lista ou de um array

```
# np.mean

b = [6,2]

print( (6+2)/2 )
print( np.mean(b) )
```

# np.var

versão resumida de *variance* ou variância e calcula a variância de uma lista ou de um array

```
# np.var

c = [3,10,100]

media_c = np.mean(c)

print( (( (3 - media_c)**2 + (10 - media_c)**2 + (100 - media_c)**2 ) )/3 )
print(np.var(c))
```

# np.std

STD é a sigla para *Standard Deviation* realiza o **desvio padrão** de todos os elementos de uma lista um de um array

```
# np.std

d = [3,10,100]

var_d = np.var(d)
print(np.sqrt(var_d))
print(np.std(d))
```

# max e min | np.max e np.min

Referente aos valores **máximos** e **mínimos** respectivamente, e retorna o valor máximo de uma lista ou array 

```
# max e min

e = [3,10,100]
print(max(e))
print(min(e))

print(np.max(e))
print(np.min(e))
```

# np.cumsum

Relativo a *cumulative sum* ou **soma cumulativa**. Retorna um array cujos elemento $i$ é a soma todos os elementos anteriores ($i = (i - 1) + (i - 2) + ... +(i - n)$). 

```
f = [1,1,10,1]

print(np.cumsum(f))
```

# np.percentile

Considerando um conjunto de dados unidimensional, a **mediana** é o termo central a estes dados, ou seja, 50% dos dados está a direita da mediana e 50% dos dados está a esquerda da mediana. Os dados podem ser subdivididos em **percentis** que seriam pequenos pacotes percentuais destes dados. 

```
g = [2,3,100,2,3]
print("média de g",np.mean(g))
print("mediana de g",np.median(g))

percen_50 = np.percentile(g,50)
percen_1 = np.percentile(g,1)

print("percentile de 50% que seria a mediana",percen_50)
print("percentile de 1%",percen_1)
```
