# Constantes

doc: [constantes](https://numpy.org/doc/stable/reference/constants)

## `np.nan`

Not a Number(NaN).

> - `np.isnan()`: Verifica se o elemento é NaN.

```
print(np.nan, np.NAN, np.NaN)
print(np.isnan(np.nan), np.isnan(3.0))
```

## `np.inf`

Infinito positivo.

> - `np.isinf()`: Verifica se o elemento é infinito positivo ou infinito negativo.
> - `np.isposinf()`: Verifica se o elemento é infinito positivo.

```
print(np.inf, np.Infinity, np.PINF, np.infty)
print(np.isinf(np.inf), np.isposinf(np.inf), np.isinf(2.0))
```

## `np.NINF`

Infinito negativo.

> - `np.isneginf()`: Verifica se o elemento é infinito negativo.

```
print(np.NINF)
print(np.isneginf(np.NINF), np.isneginf(2.0))
```

Obs:

> - `np.isfinite()`: Verifica se o elemento é finito, ou seja, não é NaN, nem infinito positivo, nem infinito negativo.

```
print(np.isfinite(np.nan), np.isfinite(np.inf), np.isfinite(np.NINF), np.isfinite(5))
```

# Arrays

## `np.linspace(start, stop, num=50)`

Cria um array com elementos igualmente espaçados.

> - `start:` Valor inicial do array.
> - `stop:` Valor final do array.
> - `num:` Quantidade de amostras do array. Padrão é 50.

doc: [np.linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html?highlight=linspace#numpy.linspace)

```
linspace = np.linspace(3, 20, num=5)
print(linspace)
```

## `np.logspace(start, stop, num=50)`

Cria um array com elementos espaçados em escala log.

> - `start:` Valor inicial do array.
> - `stop:` Valor final do array.
> - `num:` Quantidade de amostras do array. Padrão é 50.

doc: [np.logspace](https://numpy.org/doc/stable/reference/generated/numpy.logspace.html#numpy.logspace)

```
logspace = np.logspace(2, 15, num=5)
print(logspace)
```

## `np.arange(start, stop, step)`

Cria um array com elementos igualmente espaçados, com tamanho do espaço igual ao *step*.

> - `start:` Valor inicial do array. Padrão é 0.
> - `stop:` Valor final do array.
> - `step:` Tamanho do espaço entre valores adjacentes.

doc: [np.arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html#numpy.arange)



```
arange = np.arange(3, 15, 2)
print(arange)
```

## `np.zeros(shape)`

Cria um array de acordo com o tamanho informado, preenchido com zeros.

> - `shape:` Tamanho do array.

doc: [np.zeros](https://numpy.org/doc/stable/reference/generated/numpy.zeros.html?highlight=zeros#numpy.zeros)

```
zeros = np.zeros(5)
print(zeros)
zeros = np.zeros((3,3))
print(zeros)
```

## `np.ones(shape)`

Cria um array de acordo com o tamanho informado, preenchido com uns.

> - `shape:` Tamanho do array.

doc: [np.ones](https://numpy.org/doc/stable/reference/generated/numpy.ones.html#numpy.ones)

```
ones = np.ones(3)
print(ones)
ones = np.ones((3,3))
print(ones)
```

## `np.empty(shape)`

Cria um array de acordo com o tamanho informado, sem inicializar as entradas.

> - `shape:` Tamanho do array.

doc: [np.empty](https://numpy.org/doc/stable/reference/generated/numpy.empty.html#numpy.empty)

```
empty = np.empty(5)
print(empty)
empty = np.empty((3,3))
print(empty)
```

## `np.full(shape, value)`

Cria um array de acordo com o tamanho informado, preenchido com o valor informado.

> - `shape:` Tamanho do array.
> - `value:` Valor a ser preenchido.

doc: [np.full](https://numpy.org/doc/stable/reference/generated/numpy.full.html#numpy.full)

```
full = np.full(5, 3.5)
print(full)
full = np.full(5, np.nan)
print(full)
```

# Aleatórios

## `np.random.rand(shape)`

Cria um array de acordo com o tamanho informado, preenchido com valores aleatórios a partir de um distribuição uniforme entre [0, 1).

> - `shape:` Tamanho do array.

docs: [np.random.rand](https://numpy.org/doc/stable/reference/random/generated/numpy.random.rand.html?highlight=random%20rand#numpy.random.rand)

```
rand = np.random.rand(5,2)
print(rand)
```

## `np.random.randn(shape)`

Cria um array de acordo com o tamanho informado, preenchido com valores aleatórios a partir de um distribuição normal (gaussiana) de média 0 e desvio padrão 1.

> - `shape:` Tamanho do array.

docs: [np.random.randn](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randn.html?highlight=random%20randn#numpy.random.randn)

```
randn = np.random.randn(5,2)
print(randn)
```

## `np.random.choice(a, size=None)`

Gera uma amostra aleatória de um array informado

> - `a:` Array.
> - `size:` Dimensão da saída. Padrão é `None`

doc: [np.random.choice](https://numpy.org/doc/stable/reference/random/generated/numpy.random.choice.html?highlight=random%20choice#numpy.random.choice)

```
choice = np.random.choice(np.array([2.0, 3.0, 1.0, 5.0, 0.5]))
print(choice)
choice = np.random.choice(np.array([2.0, 3.0, 1.0, 5.0, 0.5]), 3)
print(choice)
choice = np.random.choice(np.array([2.0, 3.0, 1.0, 5.0, 0.5]), (3,2))
print(choice)
```

## `np.random.permutation(x)`

Permuta uma sequência (x) informada.

> - `x:` Sequência.

doc: [np.random.permutation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.permutation.html#numpy.random.permutation)

```
permutation = np.random.permutation(10)
print(permutation)
permutation = np.random.permutation([1, 3, 4, 2, 6, 10, 9])
print(permutation)
```

## `np.random.suffle(x)`

Embaralha uma sequência (x) informada.

> - `x:` Sequência.

doc: [np.random.suffle](https://numpy.org/doc/stable/reference/random/generated/numpy.random.shuffle.html#numpy.random.shuffle)

```
shuffle = np.arange(10)
print(shuffle)
np.random.shuffle(shuffle)
print(shuffle)
```

```
# Matrizes só são embaralhadas ao longo das eixo vertical (troca de linhas)
randn = np.random.randn(3,2)
print(randn)
np.random.shuffle(randn)
print(randn)
```
