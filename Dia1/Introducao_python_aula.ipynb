### Calcule o Montante mensal obtido por cada investidor.  
Dica: Evite copiar e colar pedaços de códigos feitos anteriormente por você! Para isso, vamos usar o conceito DEF para definirmos uma função que vai nos ajudar. 

```python
# Codigo aqui:
def funcao_JC(C,j,t):
    # C = capital;
    # j = taxa de juros;
    # t = tempo;
    Montante = np.zeros(t+1)
    for i in range(t+1):
        Montante[i] = C*(1+j)**i
    return Montante
```

### Atividade 8a) Defina um dicionário absolutamente vazio. Este dicionário deve conter todas as informações de um investidor específico, por exemplo, o senhor Râncio. Escreva na tela para verificação:

```python
# Codigo aqui:
Rancio = {}

# Preenchendo o rancio:
Rancio = {
    'Nome': 'Rancio',
    "Investimento": investimentos[2],
    "Taxa":taxas[2],
    "Mes": meses[2],
    "Montante": x2
}

print(Rancio["Mes"])
```

### Preencha um outro dicionário com as informações de outro investidor: Acesse a taxa fixa de juros utilizada no cálculo deste investidor:

```python
# Codigo aqui:
Placencio = {}

# Preenchendo o rancio:
Placencio = {
    'Nome': 'Placêncio',
    "Investimento": investimentos[3],
    "Taxa":taxas[3],
    "Mes": meses[3],
    "Montante": x3
}

print(Placencio["Taxa"])
```

### Atividade 8b) Agora como poderiamos criar um dicionário para armazenar todos os investidores simultaneamente?

```python
# definicao da lista de armazenamento:
dados = []
# laço para acessar cada elemento das listas pre-estabelecidas:
for nome, investimento, mes, taxa, montante in zip(nomes, investimentos, meses, taxas, montantes):
    dicio = {}
    dicio['Nome'] = nome
    dicio['Investimento'] = investimento
    dicio['Mes'] = mes
    dicio['Taxa'] = taxa
    dicio['Montante'] = montante
    dados.append(dicio)
```
