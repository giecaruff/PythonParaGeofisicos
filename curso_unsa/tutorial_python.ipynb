# Atividade 7. Cálculo do decaimento radioativo

A fórmula do decaimento radioativo é dada por:

$N(t) = N_0 \cdot e^{-kt}$

onde:
- \( N(t) \) é a quantidade de material radioativo no tempo \( t \),
- \( N_0 \) é a quantidade inicial de material radioativo,
- \( k \) é a constante de decaimento,
- \( t \) é o tempo em que o decaimento ocorre,
- \( e \) é a base do logaritmo natural (aproximadamente 2.71828).

### a) Considere que o o isótopo radioativo do urânio ($U^{238}$) possui o tempo de meia vida de 4,468 bilhões de anos. Considerando uma quantidade inicial de 1000 g de urânio, passados 65 milhões de anos, quantas gramas de urânio teremos ao final deste tempo? (lembrando que $k = \frac{ln(2)}{meia \ vida} $).

### b) Calcule o processo de decaimento do exercício a) desde o tempo 0 até o tempo de 65 milhões (obs.: faça na forma 0, 1, ... , 64, 65).


```python
## Atividade 7. a)

N0 = 1000
t2_uranio = 4.468e9

k_uranio = np.log(2) / t2_uranio
t = 65e6
Nt = N0 * np.exp(-k_uranio * t)
print(Nt, 'g')
```

```python
## Atividade 7. b)

tempo_total = 65
t2_uranio = 4.468e9
k_uranio = np.log(2) / t2_uranio
N0 = 1000

# Lembrando que no Python os índices começam com 0
N = np.zeros(tempo_total + 1)

for t in range(tempo_total + 1):
    tma = t * (10 ** 6)
    N[t] = N0 * np.exp(-k_uranio * tma)

print(N)
```

## 7. d) Quantos gramas de urânio, em média, se transformaram em chumbo a cada ano?

```python
## Atividade 7. d)

Perda = N - 1000
print(Perda)

# fazendo o calculo da media na mao!
media = 0.0
for n in Perda:
    # media += m
    media = media + n
media = media / len(Perda)
print(media, 'g')

# ================================ #

media_numpy = np.mean(Perda)
print(media_numpy, 'g')
```

## Vamos considerar agora uma rocha do início do Proterozóico (2500 Ma), contendo 1000 g de cada elemento isotópico no início do processo, e vamos visualizar um gráfico de todo o processo de decaimento.

```python
# Atividade 8. a) Definir uma função para facilitar o processo

def f_decaimento(T, t2, N0):

    k = np.log(2) / t2

    # Lembrando que no Python os índices começam com 0
    N = np.zeros(len(T)) # len(vetor_tempo) = numero de elementos neste vetor

    for t in range(len(T)): # t acessa os elementos de vetor_tempo
        tma = T[t] * (10 ** 6)
        N[t] = N0 * np.exp(-k * tma)

    return N

# ====================================================== #
# testando

vetor_tempo = np.arange(65 + 1)
N = f_decaimento(vetor_tempo, t2 = 4.468e9, N0 = 1000)
print(N)

# OBS.: Fazer o mesmo teste para o Urânio 235
```
