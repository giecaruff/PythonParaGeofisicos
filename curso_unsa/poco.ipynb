## Para facilitar a visualização, vamos criar uma função que automatize a visualização dos perfis:

```python
# Criando uma função para o plot de curvas
def plot_curves(well, depth, curves, colors, size = (12,10)):

    n_tracks = len(curves)

    fig, ax = plt.subplots(1, n_tracks, sharey=True)
    fig.set_size_inches(size)

    ax[0].set_ylabel(well.curves[depth].mnemonic + ' (' + well.curves[depth].unit + ')')
    ax[0].invert_yaxis()
    for i in range(n_tracks):
        ax[i].plot(well.curves[curves[i]].data, well.curves[depth].data, color=colors[i])
        ax[i].set_title(well.curves[curves[i]].mnemonic)
        ax[i].set_xlabel(well.curves[curves[i]].unit)
        ax[i].grid()

    plt.show()

# =========================================== #

# Plotando as curvas por meio da função criada
curves = ['CALI', 'GR', 'RHOB', 'NPHI', 'ILD']
colors = ['black','green', 'red', 'blue', 'pink']
plot_curves(well, 'DEPT', curves, colors)
```

### Volume de argila

A estimativa da argilosidade é fundamental para a avaliação petrofísica e interpretação dos dados de poços. Dentre as diversas formas de estimar a argilosidade, o cálculo do volume de argila a partir do perfil de raios gama (GR) por métodos, linear ou não-lineares, são os mais utilizados. Nessa atividade serão apresentados duas formas de realizar o cálculo do volume de argila, pelo método linear (IGR) e pelo método não-linear de Larionov.

### Linear

O volume de argila cálculado por meio do método linear é dado pela seguinte fórmula:

### $V_{SH-IGR_i} = \frac{GR_i - GR_{min}}{GR_{max} - GR_{min}}$

Onde,

> - $GR_i$ é o valor do perfil GR a cada profundidade.
> - $GR_{min}$ é valor mínimo do perfil GR e indica o menor conteúdo de argila.
> - $GR_{max}$ é valor máximo do perfil GR e indica o maior conteúdo de argila.

```python
# Calculando o volume de argila pelo método linear
gr = well.curves.GR.data
vshale_igr = (gr - min(gr)) / (max(gr) - min(gr))
print(vshale_igr)
```

Em uma situação onde os cálculos seriam realizados para diversos poços, com diferentes perfis de raios gama (GR), uma abordagem interessante seria de criar uma função para o cálculo do volume de argila apresentado anteriormente.

```python
# Criando uma função para o cálculo de volume de argila IGR
def calculate_vshale_igr(gr):

    vshale_igr = (gr - min(gr)) / (max(gr) - min(gr))

    return vshale_igr

vshale_igr = calculate_vshale_igr(well.curves.GR.data)
print(vshale_igr)
```

```python
# Adicionando a nova curva calculada nos dados de poço
igr_ci = lasio.CurveItem(mnemonic = 'VSH',data = vshale_igr, unit='-', descr='', value='')
well.curves['VSH'] = igr_ci
```

## Porosidade

Para a estimativa da porosidade geralmente são utilizados os perfis de densidade (RHOB), neutrão (NPHI) ou o sônico (DT). Nessa atividade, vamos calcular a porosidade a partir do perfil de densidade.


### Densidade

O cálculo da porosidade requer o valor da densidade da matriz da rocha, o perfil de densidade (RHOB) e a densidade do fluído presente na profundidade de interesse. A equação para o cálculo é dado por:

### $\Phi_D = \frac{\varrho_m - \varrho_b}{\varrho_m - \varrho_{fl}}$

Onde,

> - $\varrho_m$ é o valor da densidade da matriz da rocha.
> - $\varrho_b$ é valor do perfil de densidade (RHOB).
> - $\varrho_{fl}$ é valor da densidade do fluído.

```python
# Criando a função para o cálculo da porosidade a partir do perfil de densidade
def calculate_porosity_den(rhob, rhom = 2.65, rhofl = 1.10):

    phi = (rhom - rhob) / (rhom - rhofl)
    return phi

# ================================================ #
# vezes 100 para se comparar a NPHI
PHID = calculate_porosity_den(well.curves.RHOB.data)*100

phid_ci = lasio.CurveItem(mnemonic = 'PHID',data = PHID, unit='%', descr='Porosidade calculada a partir do perfil densidade', value='')

well.curves['PHID'] = phid_ci
```

## Saturação de água

A saturação de água indica a proporção de água contida nos fluídos da rocha (que além de água pode conter também hidrocarbonetos). A saturação de água, usualmente considera o perfil de resistividade (ILD) e a porosidade (PHI), mas dependendo do método, pode considerar também a argilosidade (VSH).

### Archie

O método de archie para o cálculo da saturação de água é definido por:

### $S_w ^n = \frac{A.R_w}{\phi^m + R_t}$

onde,

> - $S_w$ é a saturação de água.
> - $R_w$ é a resistividade da água.
> - $\phi$ é a porosidade.
> - $R_t$ é a resistividade medida.
> - $m$ é o expoente de cimentação.
> - $n$ é o expoente de saturação.
> - $A$ é uma constante empírica (usualmente 1)

[referência Archie AAPG](https://wiki.aapg.org/Archie_equation)

```python
# Equação de Archie:
def sw_archie(rw, rt, phi, a, m, n):

    sw = ((a*rw) / (phi**m * rt))**(1/n)
    return sw

# ================================================ #
# dividir a porosidade por 100 pois esta está em porcentagem
PHID_p = well.curves.PHID.data / 100.
SW = sw_archie(0.02, well.curves.ILD.data, PHID_p, 1, 2, 2)

# transformar todos os valores de saturação maiores ou iguais a 1 em 1
SW[SW >= 1] = 1

sw_ci = lasio.CurveItem(mnemonic = 'SW',data = SW, unit='', descr='Saturacao de agua via metodo archie', value='')

well.curves['SW'] = sw_ci
```

## Larionov

O método linear para o cálculo de argilosidade é o mais simples e tende a superestimar a argilosidade nos valores intermediários de volume de argila. Sendo assim, algumas correções são aplicadas a partir do volume IGR cálculado. O método de Larionov é um tipo de correção e pode ser dado pelas fórmulas a seguir, onde a primeira é referente as rochas recentes e a segunda as rochas mais antigas:

## recente
## (1) $V_{SH-Larionov _i} = 0.083 \times (2^{(3.7 \times V_{SH-IGR_i}) } - 1 )$

## antigas
## (2) $V_{SH-Larionov_i} = 0.33 \times (2^{(2 \times V_{SH-IGR_i}) } - 1 )$

Onde,

> - $V_{SH-IGR_i}$ é o valor do volume de argila calculado pelo método linear para cada profundidade.

```python
# vamos criar uma curva de GR sem os spikes
despike_gr = TOROK.curves.GR.data
despike_gr[despike_gr >= 105] = 105
despike_gr[despike_gr <= 55] = 55
plt.plot(despike_gr)
plt.grid()
```

```python
# Recalcular o IGR, considerando um valor máximo de 110 API (ignorando spikes)
def calculate_vshale_igr(gr):
    min = np.nanmin(gr)
    max = np.nanmax(gr)

    vshale_igr = (gr - min) / (max - min)

    return vshale_igr

# ========================================= #
# Nota: Estamos usando o Larionov old porque esta rocha pertence ao cretáceo

def calculate_vshale_lari(igr):

    vshale_lari = 0.33 * ( 2 ** (2 * igr) - 1)

    return vshale_lari

# ========================================= #

# Cálculo do igr ignorando spikes:
igr = calculate_vshale_igr(despike_gr)

# Cálculo do volume de argila
vshale = calculate_vshale_lari(igr)
vsh_ci = lasio.CurveItem(mnemonic = 'VSH',data = vshale, unit='-', descr='', value='')
TOROK.curves['VSH'] = vsh_ci

# Cálculo da porosidade
PHID = calculate_porosity_den(TOROK.curves.RHOB.data, rhom = 2.70)*100
phid_ci = lasio.CurveItem(mnemonic = 'PHID',data = PHID, unit='%', descr='', value='')
TOROK.curves['PHID'] = phid_ci

# Saturação de água
SW = sw_archie(0.02, TOROK.curves.ILD.data, PHID / 100., 1, 2, 2)
SW[SW >= 1] = 1
sw_ci = lasio.CurveItem(mnemonic = 'SW',data = SW, unit='-', descr='', value='')
TOROK.curves['SW'] = sw_ci

# Gráficos
curves = ['CALI','VSH', 'PHID', 'NPHI', 'SW']
colors = ['black','green', 'navy', 'blue', 'teal']
plot_curves(TOROK, 'DEPT', curves, colors)
```
